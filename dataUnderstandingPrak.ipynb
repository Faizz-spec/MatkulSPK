```python
import pandas as pd
import numpy as np

path = "phones_data.csv"
df = pd.read_csv(path)
df.tail(3)
df.head()
```

```python
print(df.dtypes)

df_noid = df.iloc[:, 2:]
df_noid
```

```python
df_noid.describe()
```

```python
df_noid.describe(include='all')
```

```python
# Menghitung rata-rata hanya pada kolom numerik
mean_values = df_noid.select_dtypes(include=[np.number]).mean()
print("Mean Values:")
print(mean_values)

# Menghitung total hanya pada kolom numerik
sum_values = df_noid.select_dtypes(include=[np.number]).sum()
print("\nSum Values:")
print(sum_values)
```

```python
# Seleksi hanya kolom numerik dari df_noid
df_numerik = df_noid.select_dtypes(include='number')

# Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3)
q1 = df_numerik.quantile(0.25)
q3 = df_numerik.quantile(0.75)

# Menghitung Interquartile Range (IQR)
iqr = q3 - q1

# Menampilkan hasil
iqr
```

```python
# Menyelaraskan df_noid dengan IQR
df_noid_align, iqr_new = df_noid.align(iqr, axis=1, copy=False, join='outer')

# Menghitung filter outlier berdasarkan Q1 dan Q3
outlier_filter = (df_noid_align < (q1 - 1.5 * iqr_new)) | (df_noid_align > (q3 + 1.5 * iqr_new))

# Menampilkan hasil filter outlier
outlier_filter
```

```python
# Menghitung jumlah kemunculan setiap nilai unik dalam kolom 'team_title'
team_counts = df['brand_name'].value_counts()

# Menampilkan hasil
print(team_counts)
```

```python
# Menghitung standar deviasi harga per merek smartphone
std_price = df.groupby('brand_name')['best_price'].std()

# Menampilkan hasil standar deviasi
print("Standar Deviasi Harga per Merek Smartphone:")
print(std_price)

# Menghitung rata-rata harga per merek smartphone
mean_price = df.groupby('brand_name')['best_price'].mean()

# Menampilkan hasil rata-rata
print("\nRata-rata Harga per Merek Smartphone:")
print(mean_price)
```
